```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("diamonds.csv", encoding = "ISO-8859-1") #1st column needs encoding to read pound/hashtag(#) symbol
df.head()
```

```python
#rename the Unnamed: 0
df.rename(columns={'Unnamed: 0': 'order'}, inplace =True)
```

```python
#validate
df.head()
```

```python
#set indices as order
df.set_index('order', inplace=True)
df.head()
```

```python
#make a scatterplot
sns.lmplot(x='depth', y='table', data=df)
```

```python
#add more details to scatterplot
sns.lmplot(x='depth', y='table', data=df, 
           fit_reg=False, #remove regression line
           hue='carat')   #color by evolution stage
```

```python
#boxplot showing stats for each characteristic
sns.boxplot(data=df)
```

```python
#create a dataframe containing the stats for each Pokemon
#drop Total, Stage, and Legendary
stats_df = df.drop(['price', 'carat', 'cut'], axis=1)
stats_df.head()
```

```python
#boxplot for stats
sns.boxplot(data=stats_df)
```

```python
#change size of plot
plt.subplots(figsize=(13,7))

#violin plot shows the distribution of depth for each type
sns.violinplot(x='color', y='depth', data=df)
```
