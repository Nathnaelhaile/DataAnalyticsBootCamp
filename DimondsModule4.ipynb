```python
import pandas as pd
import numpy as np
```

```python
location = "datasets/diamonds.csv"
df = pd.read_csv(location)
df.head()
```

```python
#correlation
df.corr()
```

```python
#linear regression
import statsmodels.formula.api as smf
```

```python
result = smf.ols('price ~ carat + depth + table', data=df).fit()
result.summary()
```

```python
result = smf.ols('price ~ carat + depth + table - 1', data=df).fit()
result.summary()
```
